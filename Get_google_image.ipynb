```python
import os
import requests
import json

from datetime import datetime
from datetime import date
import time
import sys
```

```python
foods = [] # Names of the images to be downloaded
base_path = "./"
API_KEY = "*****************************"
cx = "*******"
download_imgs = 200
for food in foods:
  try:
      os.mkdir(base_path+food)
  except:
      print("file exists")
  food_path=base_path+food
  count=1
  for i in range(1, download_imgs, 10):
    url = "https://www.googleapis.com/customsearch/v1?key="+API_KEY+"&cx="+cx+"&searchType=image&start="+str(i)+"&q=" + food
    try:
      res = requests.get(url)
      result = json.loads(res.text)
    except:
      print("API request failed")
    for item in result['items']:
      count = count+1
      try:
        image_url = item['link']
        img_data = requests.get(image_url).content
        img_name = food+"_"+ str(count)
        with open(food_path+"/"+ img_name + ".jpg", "wb") as handler:
            handler.write(img_data)
      except:
        print("coulndn't download image")
```
