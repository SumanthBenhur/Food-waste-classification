```python
import pandas as pd
import math
import random
import numpy as np
import plotly
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import time
from tqdm import tqdm
```

```python
def HSA(HMCR, PAR, bw, NOI): 
    arr=[]
    
    values = [32,64,128,256,512]
    rows,cols=(N,d+1)
    HM= [[0 for i in range(cols)]for j in range(rows)]





    def solve():           # calculate function value of all values in HM
        for i in range(N):
            for j in range(d):
                HM[i][2]=equation(HM[i][0],HM[i][1])


    def printHM(HM):

        for i in range(N):
            for j in range(d+1):
                print(HM[i][j],end=" ")
            print()
        

    def generate(HMCR,PAR,bw,d,N,upper,lower):
    
        for i in range(N):
            for j in range(d):
                n=random.randint(0,4)
                value = values[n]
                HM[i][j]=value
        

    def calculateNew():
        n=random.randint(0,4)
        arr.append(values[n])

        

    def getRandom():    #this will remain random.random as it needs to be compared with HMCR
        return random.random() 

    def checkPosition(arr):
        for i in range(len(HM2)):
            if(HM2[i][d]<arr[d]):
                return i
        return -1    
        

    def insert(arr,index):
        if(index==-1):
            pass
        else:
            j = N-1
            while(j!=index):
                temp=HM2[j]
                
                HM2[j] = HM2[j-1]
                HM2[j-1]=temp
                j-=1
            
            HM2[index][0]=arr[0]
            HM2[index][1]=arr[1]
            HM2[index][2]=arr[2]
            
 
            

    def checkHMCR(HMCR):
        
        for i in range(d):
            r = getRandom()
            if(r>float(HMCR)):
                calculateNew()
                
            else:
            
                temp=np.random.choice([row[i] for row in HM2])
                arr.append(temp)
                if(r <= float(PAR)):
                    
                    x = values.index(arr[i])
                    if(x<4):
                        arr[i]=values[x+bw]
                    else:
                        arr[i] = values[0]
                    
        
        fvalue=equation(arr[0],arr[1])
        arr.append(fvalue)
        fvalues.append(fvalue)
        #print("array selected is ",arr)
        
        index=checkPosition(arr)
        insert(arr,index)
        
        df=pd.DataFrame(HM2,columns=['x1','x2','Function Value'])
        function_value.append(df['Function Value'])

    generate(HMCR,PAR,bw,d,N,upper,lower) #HM
    solve()
    

    HM2=sorted(HM, key=lambda HM: HM[d],reverse=True)  


    for i in range(int(NOI)):
        # print("\nIteration number",i)
        # print("\nHM is\n")
        # printHM(HM2)
        arr.clear()
        
        checkHMCR(HMCR)
    unique_lookups.append(len(set(fvalues)))
    fvalues.clear()
    minima_per_itr.append(HM2[0])
```
